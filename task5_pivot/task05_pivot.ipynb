```python
import pandas as pd
df = pd.read_csv("../data_clean/data_merged.csv")

pivot_sensor = pd.pivot_table(
    df,
    values="value",
    index="sensor_type",
    columns="location",
    aggfunc="mean"
)

pivot_sensor
```

```python
pivot_downtime = pd.pivot_table(
    df,
    values="downtime_hours",
    index="machine_id",
    columns="issue_type",
    aggfunc="sum"
)

pivot_downtime
```

```python
stacked = pivot_downtime.stack()
stacked
```

```python
unstacked = stacked.unstack()
unstacked
```
