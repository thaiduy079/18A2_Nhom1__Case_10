```python
import pandas as pd
import numpy as np

sensor_info = pd.read_csv("../data_clean/sensor_info_clean.csv")
sensor_readings = pd.read_csv("../data_clean/sensor_readings_clean.csv")
maintenance_log = pd.read_csv("../data_clean/maintenance_log_clean.csv")
```

```python
data_sr = sensor_readings.merge(
    sensor_info,
    on="sensor_id",
    how="left"
)
```

```python
avg_value_machine_type = (
    data_sr
    .groupby(["machine_id", "sensor_type"])["value"]
    .mean()
)
```

```python
total_downtime = (
    maintenance_log
    .groupby("machine_id")["downtime_hours"]
    .sum()
)
```

```python
high_risk_machines = total_downtime.sort_values(ascending=False)
```

```python
abnormal_readings = data_sr[data_sr["value"].isna()]

abnormal_count = (
    abnormal_readings
    .groupby("machine_id")
    .size()
)
```

```python
data_sr["date"] = pd.to_datetime(data_sr["timestamp"]).dt.date
daily_avg_value = (
    data_sr
    .groupby(["machine_id", "date"])["value"]
    .mean()
)
```
