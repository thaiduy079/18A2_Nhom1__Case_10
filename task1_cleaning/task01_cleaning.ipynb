```python
import pandas as pd
import numpy as np
```

```python
sensor_info = pd.read_csv("../data_raw/sensor_info.csv")
sensor_readings = pd.read_csv("../data_raw/sensor_readings.csv")
maintenance_log = pd.read_csv("../data_raw/maintenance_log.csv")
```

```python
sensor_info["sensor_id"] = sensor_info["sensor_id"].astype(str).str.strip().str.upper()
sensor_info["machine_id"] = sensor_info["machine_id"].astype(str).str.strip().str.upper()
```

```python
sensor_info["sensor_type"] = sensor_info["sensor_type"].astype(str).str.strip().str.lower()

sensor_type_map = {
    "temperature": "Temperature",
    "temp": "Temperature",
    "vibration": "Vibration",
    "vib": "Vibration",
    "current": "Current",
    "curr": "Current"
}

sensor_info["sensor_type"] = sensor_info["sensor_type"].map(sensor_type_map)
```

```python
sensor_readings["sensor_id"] = sensor_readings["sensor_id"].astype(str).str.strip().str.upper()
sensor_readings["timestamp"] = pd.to_datetime(
    sensor_readings["timestamp"], errors="coerce"
)
sensor_readings["value"] = pd.to_numeric(
    sensor_readings["value"], errors="coerce"
)

sensor_readings.loc[sensor_readings["value"] < 0, "value"] = np.nan
sensor_readings["unit"] = sensor_readings["unit"].astype(str).str.strip().str.replace(";", "")
sensor_readings["unit"] = sensor_readings["unit"].astype(str).str.strip().str.replace(";", "")
maintenance_log["machine_id"] = maintenance_log["machine_id"].astype(str).str.strip().str.upper()

maintenance_log["maintenance_date"] = pd.to_datetime(
    maintenance_log["maintenance_date"], errors="coerce"
)
maintenance_log["issue_type"] = (
    maintenance_log["issue_type"]
    .astype(str)
    .str.strip()
    .str.lower()
    .str.replace("!", "")
)

issue_map = {
    "overheat": "Overheat",
    "vibration": "Vibration",
    "electrical": "Electrical"
}

maintenance_log["issue_type"] = maintenance_log["issue_type"].map(issue_map)
maintenance_log["downtime_hours"] = pd.to_numeric(
    maintenance_log["downtime_hours"], errors="coerce"
)
```

```python
sensor_info.to_csv("../data_clean/sensor_info_clean.csv", index=False)
sensor_readings.to_csv("../data_clean/sensor_readings_clean.csv", index=False)
maintenance_log.to_csv("../data_clean/maintenance_log_clean.csv", index=False)
```
