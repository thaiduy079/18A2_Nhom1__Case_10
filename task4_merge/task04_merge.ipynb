```python
import pandas as pd
sensor_info = pd.read_csv("../data_clean/sensor_info_clean.csv")
sensor_readings = pd.read_csv("../data_clean/sensor_readings_clean.csv")
maintenance_log = pd.read_csv("../data_clean/maintenance_log_clean.csv")
```

```python
data_merge = pd.merge(
    sensor_readings,
    sensor_info,
    on="sensor_id",
    how="left"
)
```

```python
data_merge = pd.merge(
    data_merge,
    maintenance_log,
    on="machine_id",
    how="left"
)
```

```python
unmatched_readings = data_merge[
    data_merge["maintenance_date"].isna()
]
```

```python
data_merge.to_csv(
    "../data_clean/full_sensor_data.csv",
    index=False
)
```
